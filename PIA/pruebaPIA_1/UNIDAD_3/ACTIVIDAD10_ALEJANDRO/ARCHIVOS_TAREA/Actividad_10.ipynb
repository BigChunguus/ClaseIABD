Tras la exploración inicial de los datasets, se han identificado algunos problemas en su formato que requieren ser corregidos antes de proceder con el análisis y visualización de datos.

    Problemas detectados:

Separadores incorrectos: Los archivos CSV parecen estar delimitados por ; en lugar de ,, lo que ha provocado que las columnas no se separen correctamente.

Nombres de columnas mal formateados: Los nombres de las columnas contienen caracteres especiales y descripciones largas que dificultan su uso.

Valores nulos: Existen múltiples columnas con valores completamente nulos, lo que sugiere que deben eliminarse o corregirse.

    Objetivo de este apartado

Reimportar los archivos utilizando ; como delimitador.

Eliminar columnas innecesarias o completamente vacías.


```python
import pandas as pd

# Carga de datos con separador específico
precio_vida_df = pd.read_csv('Precio_Vida.csv', sep=';')  
renta_hogar_df = pd.read_csv('Renta_por_hogar.csv', sep=';')  

# Asignación de nombres a las columnas
precio_vida_df.columns = ['Anio', 'Indice_General', 'Alimentos_y_Bebidas', 'Bebidas_Alcoolicas_y_Tabaco', 
                          'Vestido_y_Calzado', 'Vivienda_y_Energía', 'Muebles_y_Hogar', 'Sanidad', 'Transporte', 
                          'Comunicaciones', 'Ocio_y_Cultura', 'Enseñanza', 'Restaurantes_y_Hoteles', 'Otros_Bienes_y_Servicios']
renta_hogar_df.columns = ['Anio', 'Renta_Media_Cantabria']

# Eliminación de columnas vacías
precio_vida_df = precio_vida_df.dropna(axis=1, how='all')
renta_hogar_df = renta_hogar_df.dropna(axis=1, how='all')

# Vista preliminar de los datos después de la limpieza
print("Primeras filas del dataset Precio_Vida después de la limpieza:")
print(precio_vida_df.head())

print("\nPrimeras filas del dataset Renta_por_hogar después de la limpieza:")
print(renta_hogar_df.head())

# Información sobre la estructura de los datos
print("\nInformación sobre el dataset Precio_Vida después de la limpieza:")
print(precio_vida_df.info())

print("\nInformación sobre el dataset Renta_por_hogar después de la limpieza:")
print(renta_hogar_df.info())
```

# Apartado 6

En este apartado, se representará un gráfico de dispersión (scatter plot) que visualice la diferencia entre los valores de los índices en los años 2008 y 2024.

    Objetivos de este apartado:

Extraer los valores de todos los índices en los años 2008 y 2024.

Calcular la diferencia entre 2024 y 2008 para cada índice.

Representar los cambios en un scatter plot con etiquetas para cada índice.

```python
# Conversión y verificación de la columna "Anio"
precio_vida_df["Anio"] = pd.to_numeric(precio_vida_df["Anio"], errors='coerce')
print("Valores únicos en la columna 'Anio':", precio_vida_df["Anio"].unique())

# Renombrar columnas para mayor claridad
precio_vida_df.columns = ['Anio', 'Indice_General', 'Alimentos', 'Bebidas', 
                          'Vestido', 'Vivienda', 'Muebles', 'Sanidad', 
                          'Transporte', 'Comunicaciones', 'Ocio', 'Enseñanza', 
                          'Restaurantes', 'Bienes']

# Selección de datos para los años 2008 y 2024
indice_2008 = precio_vida_df[precio_vida_df["Anio"] == 2008]
indice_2024 = precio_vida_df[precio_vida_df["Anio"] == 2024]

# Lista de índices a comparar
indices = ['Indice_General', 'Alimentos', 'Bebidas', 'Vestido', 'Vivienda', 
           'Muebles', 'Sanidad', 'Transporte', 'Comunicaciones', 'Ocio', 
           'Enseñanza', 'Restaurantes', 'Bienes']

# Eliminación de valores nulos en los datos seleccionados
indice_2008 = indice_2008.dropna(subset=indices)
indice_2024 = indice_2024.dropna(subset=indices)

# Cálculo de diferencias entre 2008 y 2024
diferencias = []
for indice in indices:
    if indice in indice_2008.columns and indice in indice_2024.columns:
        diferencia = indice_2024[indice].values[0] - indice_2008[indice].values[0]
        diferencias.append(diferencia)
    else:
        diferencias.append(np.nan)

# Filtrar índices sin valores nulos
indices_filtrados = [indices[i] for i in range(len(diferencias)) if not np.isnan(diferencias[i])]
diferencias_filtradas = [d for d in diferencias if not np.isnan(d)]

# Creación del gráfico de diferencias
plt.figure(figsize=(10, 6))
plt.scatter(indices_filtrados, diferencias_filtradas, color='#8e44ad', s=100, edgecolor='black')

# Anotaciones en el gráfico
for i, indice in enumerate(indices_filtrados):
    plt.annotate(indice, (indices_filtrados[i], diferencias_filtradas[i]), 
                 textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9, rotation=45)

# Línea de referencia en 0
plt.axhline(0, color='black', linewidth=1, linestyle='--')

# Personalización del gráfico
plt.title('Diferencia en los índices entre 2008 y 2024', fontsize=14)
plt.xlabel('Índice', fontsize=12)
plt.ylabel('Diferencia (2024 - 2008)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Carga y procesamiento de datos para gráficos de tarta
try:
    df_turismo = pd.read_csv('Turismo.csv', sep=';', dtype={'Año': int})
    
    # Filtrar datos para los años 1999 y 2024
    turismo_1999 = df_turismo[df_turismo['Año'] == 1999]
    turismo_2024 = df_turismo[df_turismo['Año'] == 2024]
    
    # Definir categorías y valores
    etiquetas = ['Españoles', 'Extranjeros']
    tamaños_1999 = [turismo_1999['Residentes_España'].values[0], turismo_1999['Residentes_Extranjero'].values[0]]
    tamaños_2024 = [turismo_2024['Residentes_España'].values[0], turismo_2024['Residentes_Extranjero'].values[0]]
    
    # Creación de gráficos de tarta
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    axs[0].pie(tamaños_1999, labels=etiquetas, autopct='%1.1f%%', explode=(0.1, 0), colors=['lightblue', 'orange'])
    axs[0].set_title('Distribución de turistas (1999)')
    
    axs[1].pie(tamaños_2024, labels=etiquetas, autopct='%1.1f%%', explode=(0.1, 0), colors=['lightblue', 'orange'])
    axs[1].set_title('Distribución de turistas (2024)')
    
    plt.show()

except Exception as e:
    print(f"Error al cargar los datos: {e}")
```

# Apartado 10

    En este apartado, utilizaremos Seaborn para analizar la relación entre la tarifa del boleto (fare) y tres variables clave en el dataset Titanic:

Clase del pasajero (pclass) con un gráfico de caja (boxplot).

Supervivencia (survived) con un gráfico de dispersión tipo enjambre (swarmplot).

Puerto de embarque (embark_town) con un gráfico de distribución proporcional (displot) ECDF.

    Carga y visualización de los datos

A continuación, crearemos una figura con tres gráficos distintos dentro de la misma visualización.

```python
# Análisis de la relación entre tarifa y variables clave en el Titanic
try:
    df_titanic = sns.load_dataset('titanic')
    
    # Definir paletas de colores personalizadas para los gráficos
    paleta_boxplot = ['#f55353', '#4bc8ed', '#58f076']  # Colores para el boxplot
    paleta_swarmplot = {0: '#aa7fdb', 1: '#ea9d3f'}  # Colores para el swarmplot
    paleta_ecdf = ['#890cf7', '#146414', '#f70c8c']  # Colores para el ecdfplot
    
    # Crear subgráficos para mostrar las relaciones
    fig, ejes = plt.subplots(1, 3, figsize=(18, 6))
    
    # Boxplot para mostrar la distribución de tarifa por clase
    sns.boxplot(data=df_titanic, x='pclass', y='fare', ax=ejes[0], palette=paleta_boxplot)
    ejes[0].set_title('Distribución de Tarifa por Clase')
    
    # Swarmplot para mostrar la tarifa según supervivencia
    sns.swarmplot(data=df_titanic, x='survived', y='fare', ax=ejes[1], hue='survived', palette=paleta_swarmplot, legend=False)
    ejes[1].set_title('Tarifa por Supervivencia')
    
    # ECDF para mostrar la distribución de tarifas por puerto de embarque
    sns.ecdfplot(data=df_titanic, x='fare', hue='embark_town', ax=ejes[2], palette=paleta_ecdf)
    ejes[2].set_title('Distribución Proporcional de Tarifa por Puerto de Embarque')
    
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Error en el análisis del Titanic: {e}")
```

# ¿Qué conclusiones podemos tomar de los datos del df si observamos la primera gráfica?
La primera gráfica es un boxplot que muestra la distribución de las tarifas (fare) en función de la clase del pasajero (pclass).

Se puede observar que los pasajeros de primera clase pagaron tarifas significativamente más altas en comparación con los de segunda y tercera clase.
La mediana de la tarifa es más alta en primera clase, y hay una mayor dispersión de valores, lo que sugiere que algunos pasajeros pagaron tarifas extremadamente altas.
En tercera clase, las tarifas son mucho más bajas y hay menos variabilidad en los precios.
En conclusión, la clase del pasajero tenía una gran influencia en el costo del boleto, con precios más elevados para los pasajeros de mayor estatus.
# ¿Qué problema encontramos con la segunda?
La segunda gráfica es un swarmplot, que muestra la relación entre la tarifa y la supervivencia. Sin embargo, hay un problema importante:

Superposición de puntos: Dado que los datos contienen muchas observaciones con valores similares, los puntos pueden amontonarse demasiado y hacer difícil distinguir patrones claros.
Escalado de valores extremos: Como las tarifas varían en un rango amplio, algunos valores muy altos pueden hacer que los puntos más bajos sean menos visibles.
Advertencia de Seaborn: A partir de versiones recientes, swarmplot puede generar advertencias cuando hay muchos puntos, lo que puede afectar el rendimiento.
Posible solución: Se podría usar un violin plot o un jitter plot con sns.stripplot() para mejorar la visualización sin tanta superposición.
# ¿Qué ciudad es más cara para adquirir el boleto?
La tercera gráfica (ECDF) nos muestra la distribución proporcional de las tarifas en función del puerto de embarque (embark_town). Para determinar qué ciudad tenía los boletos más caros, podemos observar cuál tiene la mayor proporción de tarifas altas.

Generalmente, en el dataset del Titanic:

Cherburgo (C) tiende a tener las tarifas más altas, ya que muchos pasajeros de primera clase abordaron en este puerto.
Southampton (S) tiene una distribución más equilibrada con muchas tarifas medias y bajas.
Queenstown (Q) tiene las tarifas más bajas en general, con la mayoría de los pasajeros en tercera clase.
Por lo tanto, Cherburgo es la ciudad más cara para adquirir el boleto.

# Apartado 14

En este apartado, utilizaremos Seaborn para analizar la relación entre la edad de los pasajeros, la tarifa que pagaron y la clase en la que viajaban, incorporando además el número de familiares a bordo (sibsp) como una variable adicional.

    Visualización: Crearemos un gráfico conjunto (jointplot) con sns.jointplot, donde:

Eje X: Representará la edad (age) del pasajero.

Eje Y: Mostrará la tarifa (fare) que pagaron.

Hue (pclass): Diferenciará a los pasajeros según su clase (1ª, 2ª y 3ª).

Se visualizarán distribuciones marginales en los ejes X e Y.

Este gráfico nos permitirá explorar la distribución de tarifas en función de la edad, clase y el número de familiares a bordo, para detectar posibles patrones de viaje en el Titanic.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
df_datos = sns.load_dataset('titanic')

# Eliminar filas con valores nulos en las columnas relevantes
df_datos = df_datos.dropna(subset=['age', 'fare', 'pclass'])

# Crear el gráfico conjunto
grafico = sns.jointplot(
    data=df_datos, x="age", y="fare", hue="pclass", 
    palette=["#d4a4a4", "#a48ca4", "#4a2c4a"],  # Colores personalizados
    kind="scatter", alpha=0.6, marginal_kws={"common_norm": False}
)

# Ajustar los gráficos marginales según la clase
for clase, color in zip([1, 2, 3], ["#d4a4a4", "#a48ca4", "#4a2c4a"]):
    subconjunto = df_datos[df_datos["pclass"] == clase]  # Filtrar por clase
    sns.kdeplot(data=subconjunto, x="age", fill=True, ax=grafico.ax_marg_x, alpha=0.4, color=color)
    sns.kdeplot(data=subconjunto, y="fare", fill=True, ax=grafico.ax_marg_y, alpha=0.4, color=color)

# Ajustes finales
grafico.ax_joint.set_xlim(-5, 85)  
grafico.ax_joint.set_ylim(-10, 600)  
grafico.ax_joint.set_xlabel("Edad", fontsize=12)
grafico.ax_joint.set_ylabel("Tarifa", fontsize=12)
grafico.figure.suptitle("Relación entre Edad, Tarifa y Clase en el Titanic", fontsize=14)

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import plotly.express as px
from plotly.offline import init_notebook_mode

# Habilitar renderizado interactivo en Jupyter
init_notebook_mode(connected=True)

# Cargar los datos desde el archivo CSV
archivo_csv = "GlobalWeatherRepository.csv"
df = pd.read_csv(archivo_csv)

# Convertir la columna "last_updated" a tipo datetime
df["last_updated"] = pd.to_datetime(df["last_updated"])

# Filtrar solo los países deseados
paises_interes = ["Spain", "Thailand", "New Zealand", "United States of America"]
df_filtrado = df[df["country"].isin(paises_interes)]

# Agrupar por país y fecha (cada día), luego calcular la temperatura media
df_grouped = df_filtrado.groupby(["country", "last_updated"])["temperature_celsius"].mean().reset_index()

# Crear el gráfico de líneas interactivo con Plotly
fig = px.line(
    df_grouped, 
    x="last_updated", 
    y="temperature_celsius", 
    color="country", 
    markers=True, 
    text="temperature_celsius", 
    title="Temperatura Media Diaria por País",
    labels={"temperature_celsius": "Temperatura Media (°C)", "last_updated": "Fecha", "country": "País"}
)

# Ajustar la posición del texto para que quede un poco más arriba
fig.update_traces(
    textposition="top center",  # Mantenerlo arriba del punto
    mode="lines+markers+text",  # Mostrar líneas, puntos y texto
)

# Personalización del gráfico (ejes y leyenda)
fig.update_layout(
    xaxis_title="Fecha", 
    yaxis_title="Temperatura Media (°C)", 
    legend_title="País"
)

# Mostrar el gráfico
fig.show()
```

Todas las figuras usarán el template:     template="plotly_dark",


    Integra todas las figuras en un dashboad con la librería DASH: from dash import Dash, dcc, html. En la primera fila se encontrará sólo el mapa del mundo y en la segunda ambas gráficas.


    Personaliza el dashboard para que:

```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html

# Cargar los datos
archivo_csv = "TemperaturasGlobales.csv"
df = pd.read_csv(archivo_csv)

# 1) Gráfico de líneas: Evolución de la temperatura media global por año
# Seleccionar las columnas de los años
years = [str(year) for year in range(1970, 2022)]
df_temp_global = df[years].mean(axis=0).reset_index()
df_temp_global.columns = ["Año", "Temperatura Media Global (°C)"]

# Crear el gráfico de líneas
fig1 = px.line(
    df_temp_global,
    x="Año",
    y="Temperatura Media Global (°C)",
    title="Evolución de la Temperatura Media Global por Año",
    labels={"Año": "Año", "Temperatura Media Global (°C)": "Temperatura Media Global (°C)"},
    template="plotly_dark",
    line_shape="spline"
)

# Personalizar color de la línea
fig1.update_traces(line=dict(color="red"))

# 2) Gráfico de barras: Top 10 países con mayor cambio de temperatura
# Calcular el cambio de temperatura como la diferencia entre el valor máximo y mínimo por país
df['Temperature Change'] = df[years].max(axis=1) - df[years].min(axis=1)

# Seleccionar los 10 países con mayor cambio de temperatura
df_top10 = df[['Country Name', 'Temperature Change']].nlargest(10, 'Temperature Change')

# Crear el gráfico de barras
fig2 = px.bar(
    df_top10,
    x="Country Name",
    y="Temperature Change",
    title="Top 10 Países con Mayor Cambio de Temperatura",
    labels={"Country Name": "País", "Temperature Change": "Cambio de Temperatura (°C)"},
    template="plotly_dark",
    color="Temperature Change",
    color_continuous_scale="reds"
)

# 3) Mapa en forma de globo con la variación de temperatura en 2021 (solo positivos)
# Calcular la variación de temperatura en 2021 (solo valores positivos)
df['Temperature Change 2021'] = df['2021'] - df['1970']

# Filtrar para considerar solo los valores positivos
df_2021_positive = df[df['Temperature Change 2021'] > 0]

# Crear el gráfico del mapa
fig3 = px.choropleth(
    df_2021_positive,
    locations="Country Name",
    locationmode="country names",
    color="Temperature Change 2021",
    hover_name="Country Name",
    title="Variación de Temperatura en 2021 (Solo Valores Positivos)",
    template="plotly_dark",
    color_continuous_scale="reds",
    projection="orthographic"  # Proyección en forma de globo
)

# Crear la app de Dash
app = Dash(__name__)

app.layout = html.Div(style={'backgroundColor': '#111111', 'color': 'white', 'padding': '20px'}, children=[

    html.H1("Dashboard de Evolución de Temperaturas Globales", style={'textAlign': 'center'}),

    # Primera fila: Mapa de variación de temperatura en 2021 (solo positivos)
    html.Div(children=[dcc.Graph(figure=fig3)]),

    # Segunda fila: Gráfico de líneas y barras
    html.Div(style={'display': 'flex', 'justify-content': 'space-around'}, children=[
        dcc.Graph(figure=fig1, style={'width': '48%'}),
        dcc.Graph(figure=fig2, style={'width': '48%'})
    ])
])

# Ejecutar el servidor
if __name__ == '__main__':
    app.run_server(debug=True)
```
