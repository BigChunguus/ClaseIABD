### ** Regresión Lineal **

1. Buscamos la linea que se ajuste mejor a la relación que hay entre las variables de entrada y salida

```python
%pip install pandas numpy matplotlib sklearn

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

2. Tomamos como DF la relación entre el precio y el área de unos terrenos

```python
df = pd.DataFrame()
df["Areas"] = [2600, 3000, 3200, 3600, 4000]
df["Precios"] = [550000, 565000, 610000, 680000, 725000]

plt.scatter(df["Areas"], df["Precios"])
```

3. Establecemos los conjuntos de datos

```python
# Variable de entrada independiente: es un dataframe que suele ser multidimensional, y por eso se escribe en mayúscula
X = df[["Areas"]]
type(df[["Areas"]])
```

```python
# Variable de salida dependiente:
y = df["Precios"]
```

4. Creamos y entregamos el modelo

```python
modelo = linear_model.LinearRegression()
modelo.fit(X,y)
```

5. Probamos el modelo

```python
modelo.predict([[3000]])
```

```python
plt.plot(df["Areas"], modelo.predict(X))
plt.scatter(df["Areas"], df["Precios"])
```

6. Evaluamos el modelo

```python
# Importamos las librerias de sklearn
from sklearn.metrics import mean_squared_error as msef
from sklearn.metrics import mean_absolute_error as maef
from sklearn.metrics import r2_score as r2f

# Creamos una variable con todas las predicciones
y_pred = modelo.predict(X)

# Calculamos métricas
mse = msef(y, y_pred)
rmse = np.sqrt(mse)
mae = maef(y, y_pred)
r2 = r2f(y, y_pred)

# Mostramos los valores
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R2: {r2:.2f}")
```
