2. Importar y exportar archivos

```python
# ¿Y si queremos obtener los índices de los valores minimos y máximos?
min_id = df['2022'].idxmin()
max_id = df['2022'].idxmax()

munimin = df.loc[min_id, 'Municipio']
munimax = df.loc[max_id, 'Municipio']
munimin, munimax
```

```python
# Seleccionar columnas con una condición
# Por ejemplo, municipios con un consumo en 2022 mayor de 100000kW
df[df['2022'] > 100000]
```

```python
# Queremos seleccionar solo las columnas de los años (suponemos que son todas menos la del Municipio)
# Y luego calcular su media
df['Media Consumo'] = df.iloc[:, 2:].mean(axis=1)
df
```
