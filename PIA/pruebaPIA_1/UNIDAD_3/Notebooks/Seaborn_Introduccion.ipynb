1. Ejemplo de la web

```python
sns.set_theme(style="ticks")

rs = np.random.RandomState(11)
x = rs.gamma(12, size=1000)
y = -.5 * x + rs.normal(size=1000)

sns.jointplot(x=x, y=y, kind="hex", color="#ea97ff")
```

2. Aplicar otro ejemplo en este caso utilizando un dataset prediseñado de Seaborn

```python
# Aplicar un tema por defecto
sns.set_theme()

# Cargamos un dataset de ejemplo
propinas = sns.load_dataset("tips")

propinas
```

```python
# Crear una visualización mediante el método replot: sirve para visualizar la relación entre dos variables
sns.relplot(
    data= propinas,
    x = "total_bill",
    y = "tip",
    col = "time",
    hue = "sex",
    style = "sex",
    size = "size"
)
```

3. Cargamos nuestro porpio CSV y generamos un dataframe

```python
# Utilizaremos seaborn para simplificar la creación de gráficos y estilizar
# Utilizaremos matplotlib para controlar la personalización
# Utilizaremos pandas para cargar los DataFrames

# Cargamos el csv 
ruta = "Lluvias_region_A.csv"
df = pd.read_csv(ruta)
df.head()
```

```python
# Configurar el estilo de seaborn
sns.set_style(style="darkgrid")

# Crear un gráficos de barras (precipitaciones por región)
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='region', y='anual', palette='viridis');

plt.title("Precipitaciones anuales por region", fontsize = 16)
plt.xlabel("Region")
plt.ylabel("Precipitaciones anuales (mm)")
plt.xticks(rotation=45)

plt.tight_layout()
```

```python
sns.relplot(data = propinas,
            x = "total_bill",
            y = "tip",
            kind = "line")
```

5. Representación Distributiva

```python
# Importamos otra libreria de seaborn
pinguinos = sns.load_dataset("penguins")
pinguinos
```

```python
sns.displot(data = pinguinos,
            x = "body_mass_g",
            kde = True,
            col = "sex"
            )
```

6. Distribución categorica

```python
# Gráfico de dispersión
# Los puntos que se dibujan para cada categoría lo hacen a lo largo de un eje, y su ajuste horizontal es aleatorio
sns.catplot(data=propinas,
            kind="strip",
            x="day",
            y="total_bill",
            hue="sex")
```

```python
#Tipo swarmplot: los valores no se superponen
sns.catplot(data=propinas,
            kind="swarm",
            x="day",
            y="total_bill",
            hue="sex")
```

```python
# Diagrama de caja
sns.catplot(data=propinas,
            kind="box",
            x="day",
            y="total_bill",
            hue="sex")
```

```python
# Creamos una figura plot
plt.figure(figsize=(12,6))
sns.kdeplot(data=propinas, x="tip", alpha=0.6, fill=True, color="purple")
sns.boxplot(data=propinas, x="tip", orient="h", color="orange", width=0.2)
```

```python
# Diagrama de violin
sns.catplot(data=propinas,
            kind="violin",
            x="day",
            y="total_bill",
            hue="sex")
```

```python
# Diagrama de punto
sns.catplot(data=propinas,
            kind="point",
            x="day",
            y="total_bill",
            hue="sex")
```

```python
# Diagrama de barras
sns.catplot(data=propinas,
            kind="bar",
            x="day",
            y="total_bill",
            hue="sex")
```
