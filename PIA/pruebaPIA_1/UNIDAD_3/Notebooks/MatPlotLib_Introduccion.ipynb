1. Creamos un gr√°fico con plt.plot()

```python
a = [1,6,8,3,2,243,65,32,1,23,31]
plt.plot(a); # el ; elimina la informaci√≥n innecesaria
```

```python
# Vamos a crear dos listas y graficarlas
x = list(range(131))
y = []
for numero in x:
    y.append(numero**2)

plt.figure(facecolor='slateblue')
plt.plot(x,y)
```

```python
# Podemos tambi√©n cambiar el formato
x_0 = list(range(10))
y_0 = [2,3,1,6,9,7,8,8,5,3]
plt.plot(x_0, y_0, 'b*--')
```

```python
# Gr√°ficos con el m√©todo Orientado a Objetos
# fig: todos los elementos de la figura
# ax: los ejes
fig, ax = plt.subplots()

# Personalizar el gr√°fico a√±adiendo un titulo al gr√°fico y a los ejes
ax.set(title='Gr√°fico x^2', xlabel='Valores de X', ylabel='Valores de Y')
ax.plot(x,y)

fig.savefig('ejemplo_x^2.png')
```
